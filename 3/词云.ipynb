```python
import pandas as pd
import matplotlib.pyplot as plt
from wordcloud import WordCloud
```

```python
basket = pd.read_csv('./Market_Basket_Optimisation.csv', header=None)

# 生成词云

print('正在根据词频生成词云......')

text = []

for _, value in basket.fillna('').iteritems():
    text.extend(value)

cut_text = ','.join(text)

wc = WordCloud(
    max_words=50,
    width=2000,
    height=1200,
)
wordcloud = wc.generate(cut_text)

plt.imshow(wordcloud)
plt.axis('off')
plt.show()

```
